```acl2
(defun bounded-recursive-eval (exp n)
  (declare (xargs :guard (natp n)
                  :measure (nfix n)))
  (if (zp n) exp (bounded-recursive-eval (ev$ exp '()) (1- n))))
```

```acl2
(thm (equal (bounded-recursive-eval *quine-exp* N) *quine-exp* ))
```
